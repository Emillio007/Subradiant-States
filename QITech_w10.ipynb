Exercise 2a: plot of number of extractable bits by Hmin vs. von Neumann extractor:

```python
import numpy as np
import matplotlib.pyplot as plt

q = np.linspace(0, 1, 1000)
eps = 1

vN = lambda q: q * (1 - q)

def hmin(q):
    if q < 0.5:
        return -np.log2(1-q)
    elif q > 0.5:
        return -np.log2(q)
    else:
        return -np.log2(0.5)

hminny = []
for val in q:
    hminny.append(hmin(val))

hminN = hminny - 2*np.log2(1/eps)

y_vN = vN(q)

plt.plot(q, y_vN, label="vN")
plt.plot(q, hminN, label="Hmin, eps=%s" %eps)
plt.xlabel("q")
plt.ylabel("Probability of extracting bit")
plt.legend()
```
