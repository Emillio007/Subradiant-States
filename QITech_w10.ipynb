Exercise 2a: plot of number of extractable bits by Hmin vs. von Neumann extractor:

```python
import numpy as np
import matplotlib.pyplot as plt

q = np.linspace(0, 1, 1000)

vN = lambda q: q * (1 - q)

def hminN(q):
    if q < 0.5:
        return q
    elif q > 0.5:
        return 1 - q
    else:
        return 0.5

y_hminN = []
for val in q:
    y_hminN.append(hminN(val))

y_vN = vN(q)

plt.plot(q, y_vN, label="vN")
plt.plot(q, y_hminN, label="Hmin")
plt.xlabel("q")
plt.ylabel("Probability of extracting bit")
plt.legend()
```
